```python
# agrupando mortes por pa√≠s
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1) # juntando e criando colunas
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

### Verificando a Taxa de Letalidade

Tratar de letalidade durante uma epidemia √© algo dificil, pois √© algo que est√° acontecendo ainda, o que complica na hora de monitorar.

```python
combined.sum()
```

```python
combined['deaths'].sum()
```

```python
letality = combined['deaths'].sum() / combined['confirmed'].sum()
print((letality * 100).round(2))
```

```python
combined['letality_rate1'] = combined['deaths'] / combined['confirmed'] * 100 # criando a coluna de letalidade
combined.head()
```

Aqui vamos ter apenas valores 0, pelo fato do conjunto de dados est√° com problema na `recovered`, assim trazendo problemas aos c√°lculos.

```python
combined['letality_rate2'] = combined['deaths'] / (combined['deaths']  + combined['recovered']) * 100
```

```python
combined.head()
```

```python
combined.sort_values('letality_rate1', ascending=False) # ordenando a maior letalidade
```

## Taxa de Incid√™ncia

Essa taxa √© definida como o n√∫mero de casos novos de uma doen√ßa ou outra afec√ß√£o de sa√∫de dividido pela popula√ß√£o em risco da doen√ßa (popula√ß√£o exposta) em um espa√ßo geogr√°fico durante um tempo especificado. Defini√ß√£o da `OPAS`.

```python
cases = confirmed.groupby('is_china_label').sum()[last_date].loc['China'] # pegando apenas caso na china
estimated_population = 1453144312 # popula√ß√£o da china aproximadamente
incidence_ratio = cases / estimated_population * 100000 # incidencia por 100k de habitantes
incidence_ratio 
```

```python
# acessando os dados da UN apenas com os links deles
un_population = pd.read_csv('https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2022_TotalPopulationBySex.zip')
un_population = un_population[['LocID', 'Location', 'VarID', 'Variant', 'Time',	'MidPeriod',	'PopMale',	'PopFemale',	'PopTotal']]
un_population.head()
```

Os dados acima foram pegos do site da UN, podemos baixar e importar para o notebook ou podemos acessar direto pelo link, fica a crit√©rio, que pode ser acesado pelo seguinte link : https://population.un.org/wpp/Download/Standard/CSV/

```python
# achar pa√≠ses especificios
def locate(population, country):
  return un_population.query(f'Location.str.contains("{country}") and Time == 2021', engine='python')

locate(un_population, 'Rus')
```

```python
# para os casos com o indice diferente, no caso nome dos pa√≠ses
# aqui estou mudando united state of america para US igual est√° na Jonh Hop...
def rename_location(location):
  if location == 'United States of America': # posso colocar mais if e ir fazendo para outros pa√≠ses como Korea, Iran, etc
    return 'US'
  if location == 'Republic of Korea':
    return 'Korea, South'
  if location == 'Russian Federation':
    return 'Russia'
  return location

un_population_total['location_for_who'] = un_population_total.index.map(rename_location) # map √© a nossa fun√ß√£o para aplicar em todo o conj. de dados
un_population_total.query('location_for_who == "Russia"').head()
# un_population_total.query('location_for_who == "China"')
```

```python
combined_expanded = combined.join(un_population_total.set_index('location_for_who')).sort_values('confirmed', ascending=False).dropna()
combined_expanded.head()
```

**Aqui vamos calcular a taxa de incid√™ncia**

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000 # reportamos a cada 100k de habitantes, por isso o * 100000
combined_expanded.head()
```

Vale lembrar que a taxa de incid√™ncia √© com base no que foi confirmado, ent√£o caso tenha omiss√£o de dados nos confirmados a taxa ter√° sua altera√ß√£o. Por isso a grande necessidade de testar a popula√ß√£o para se ter no√ß√µes reais da gravidade da situa√ß√£o, por√©m muitos pa√≠ses omitiram certas informa√ß√µes. N√£o podemos tamb√©m dizer que √© s√≥ omiss√£o, √†s vezes o pa√≠s reagiu r√°pido e impediu que se tivesse um aumento de pessoas infectadas. 

```python
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head()
```
