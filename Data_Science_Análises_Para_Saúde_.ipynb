```python
# agrupando mortes por país
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1) # juntando e criando colunas
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

### Verificando a Taxa de Letalidade

Tratar de letalidade durante uma epidemia é algo dificil, pois é algo que está acontecendo ainda, o que complica na hora de monitorar.

```python
combined.sum()
```

```python
combined['deaths'].sum()
```

```python
letality = combined['deaths'].sum() / combined['confirmed'].sum()
print((letality * 100).round(2))
```

```python
combined['letality_rate1'] = combined['deaths'] / combined['confirmed'] * 100 # criando a coluna de letalidade
combined.head()
```

Aqui vamos ter apenas valores 0, pelo fato do conjunto de dados está com problema na `recovered`, assim trazendo problemas aos cálculos.

```python
combined['letality_rate2'] = combined['deaths'] / (combined['deaths']  + combined['recovered']) * 100
```

```python
combined.head()
```

```python
combined.sort_values('letality_rate1', ascending=False) # ordenando a maior letalidade
```

```python
# um exemplo de má visualização

combined[['confirmed', 'letality_rate1']].plot()
```

```python
import seaborn as sns

sns.scatterplot(data=combined, x='confirmed', y='letality_rate1')
```

```python
sns.scatterplot(data=combined.query("confirmed > 40000"), x='confirmed', y='letality_rate1') # quanto maior o número de casos no país, percebemos uma diminuição da letalidade
```

```python
sns.distplot(combined.query('confirmed > 400000')['letality_rate1'])
```

```python
# filtrando e verificando a letalidade
combined.sort_values('letality_rate1', ascending=False).query('confirmed > 4000').head(10)
```

## Taxa de Incidência

Essa taxa é definida como o número de casos novos de uma doença ou outra afecção de saúde dividido pela população em risco da doença (população exposta) em um espaço geográfico durante um tempo especificado. Definição da `OPAS`.

```python
cases = confirmed.groupby('is_china_label').sum()[last_date].loc['China'] # pegando apenas caso na china
estimated_population = 1453144312 # população da china aproximadamente
incidence_ratio = cases / estimated_population * 100000 # incidencia por 100k de habitantes
incidence_ratio 
```

Só que aqui estamos usando a população total da China e não a população que está sendo afetada, no caso de cada província.

```python
# acessando os dados da UN apenas com os links deles
un_population = pd.read_csv('https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2022_TotalPopulationBySex.zip')
un_population = un_population[['LocID', 'Location', 'VarID', 'Variant', 'Time',	'MidPeriod',	'PopMale',	'PopFemale',	'PopTotal']]
un_population.head()
```

Os dados acima foram pegos do site da UN, podemos baixar e importar para o notebook ou podemos acessar direto pelo link, fica a critério, que pode ser acesado pelo seguinte link : https://population.un.org/wpp/Download/Standard/CSV/

```python
# achar países especificios
def locate(population, country):
  return un_population.query(f'Location.str.contains("{country}") and Time == 2021', engine='python')

locate(un_population, 'Rus')
```

```python
# para os casos com o indice diferente, no caso nome dos países
# aqui estou mudando united state of america para US igual está na Jonh Hop...
def rename_location(location):
  if location == 'United States of America': # posso colocar mais if e ir fazendo para outros países como Korea, Iran, etc
    return 'US'
  if location == 'Republic of Korea':
    return 'Korea, South'
  if location == 'Russian Federation':
    return 'Russia'
  return location

un_population_total['location_for_who'] = un_population_total.index.map(rename_location) # map é a nossa função para aplicar em todo o conj. de dados
un_population_total.query('location_for_who == "Russia"').head()
# un_population_total.query('location_for_who == "China"')
```

```python
combined_expanded = combined.join(un_population_total.set_index('location_for_who')).sort_values('confirmed', ascending=False).dropna()
combined_expanded.head()
```

**Aqui vamos calcular a taxa de incidência**

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000 # reportamos a cada 100k de habitantes, por isso o * 100000
combined_expanded.head()
```

Vale lembrar que a taxa de incidência é com base no que foi confirmado, então caso tenha omissão de dados nos confirmados a taxa terá sua alteração. Por isso a grande necessidade de testar a população para se ter noções reais da gravidade da situação, porém muitos países omitiram certas informações. Não podemos também dizer que é só omissão, às vezes o país reagiu rápido e impediu que se tivesse um aumento de pessoas infectadas. 

```python
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 100000
combined_expanded.head()
```
